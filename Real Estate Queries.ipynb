```python
# Query 1 - Retrieve all distinct towns from the table.

query_1 = """
SELECT DISTINCT(town) FROM real_estate;
"""
```

```python
# Query 2 - Find the total number of properties for each state.

query_2 = """
SELECT state, COUNT(title) AS number_of_properties FROM real_estate
GROUP BY state;
"""
```

```python
# Query 3 - Calculate the average price of properties in each town.

query_3 = """
SELECT town, AVG(price) AS avg_price_of_properties FROM real_estate
GROUP BY town;
"""
```

```python
# Query 6 - Count the number of properties with parking space in each town.

query_6 = """
SELECT town, COUNT(title) FROM real_estate
WHERE parking_space > 0
GROUP BY town;
"""
```

```python
# Query 7 - Find the title and price of the cheapest property in each state.

query_7 = """
WITH CheapestProperties AS (
    SELECT
        title,
        price,
        state,
        ROW_NUMBER() OVER (PARTITION BY state ORDER BY price ASC) AS rank_within_state
    FROM
        real_estate
)
SELECT
    title,
    price,
    state
FROM
    CheapestProperties
WHERE
    rank_within_state = 1;
"""
```

```python
# Execute the query

cur.execute(query_7)
```

```python
# Print out the result of the query

cheapest_house_per_state = cur.fetchall()
cheapest_house_per_state
```

```python
# Query 8 - Calculate the total number of bedrooms and bathrooms in each town.

query_8 = """
SELECT town, (SUM(bedrooms) + SUM(bathrooms)) AS num_of_bedrooms_and_bathrooms FROM real_estate
GROUP BY town;
"""
```

```python
# Query 10 - Find the properties with the highest price in each state.

query_12 = """
WITH RankedProperties AS (
    SELECT title, price, state,
        ROW_NUMBER() OVER (PARTITION BY state ORDER BY price DESC) AS rank_within_state
    FROM real_estate
)
SELECT title, price, state
FROM RankedProperties
WHERE rank_within_state = 1;
"""
```

```python
# Execute the query

cur.execute(query_10)
```

```python
# Print out the result of the query

result = cur.fetchall()
result
```

```python
# Query 11 - Calculate the average price of properties with parking space and without parking space.

query_13 = """
SELECT
    AVG(CASE WHEN parking_space > 0 THEN price END) AS avg_price_with_parking,
    AVG(CASE WHEN parking_space <= 0 THEN price END) AS avg_price_without_parking
FROM
    real_estate;
"""
```

```python
# Execute the query

cur.execute(query_11)
```

```python
# Print out the result of the query

result_1 = cur.fetchall()
result_1
```

```python
# Query 12 - Retrieve the properties with the same number of bedrooms, bathrooms, and toilets.

query_12 = """
SELECT title, bedrooms, bathrooms, toilets
FROM real_estate
WHERE
    (bedrooms, bathrooms, toilets) IN (
        SELECT bedrooms, bathrooms, toilets
        FROM real_estate
        GROUP BY bedrooms, bathrooms, toilets
        HAVING COUNT(*) > 1
    );
"""
```

```python
# Execute the query

cur.execute(query_12)
```

```python
# Print out the result of the query

result_2 = cur.fetchall()
result_2
```

```python
# Query 13 - Find the properties with the highest total number of rooms (bedrooms + bathrooms + toilets) in each town.

query_15 = """
WITH RankedProperties AS (
    SELECT title, bedrooms, bathrooms, toilets, town,
        ROW_NUMBER() OVER (PARTITION BY town ORDER BY (bedrooms + bathrooms + toilets) DESC) AS rank_within_town
    FROM
        real_estate
)
SELECT title, bedrooms, bathrooms, toilets, town
FROM RankedProperties
WHERE rank_within_town = 1;
"""
```

```python
# Execute the query

cur.execute(query_13)
```

```python
# Print out the result of the query

result_3 = cur.fetchall()
result_3
```
